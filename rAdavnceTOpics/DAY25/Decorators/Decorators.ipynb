# Decorators

A decorator in Python is a function that modifies another function (or method or class) without changing its source code.
In simpler words :decortors are the functions that take the argument an another function,and add some functionality (decoration) to it then returns that function.

```python
# WHen our functions usually accept arguments. 
# So decorators need a way to accept and forward 
# those arguments.

# That's where *args and **kwargs come in.
# for example
def My_decorator(funct):
    
    def wrapper(*args,**kwargs):
        print(*args,*kwargs.items())
    return wrapper

@My_decorator
def show(*args,**kwargs):
    pass

show(12,32,12,name="hello",age=18) # agument given
```

```python
# TIMING Decorator
# timer Decorator of any function

import time
def Execution_time(funct):
    def wrapper(*args):
        start=time.time()
        funct(*args)
        end=time.time()
        print('TIme Taken By the Function is ',funct.__name__,round(end-start,5),"secs")
    return wrapper

@Execution_time
def Square(list):
    Square=[int(i)**2 for i in list]
    print(Square)
    time.sleep(2) # to get More TIme For Better view
    return Square

mylist=[i for i in range(0,100000)]
Square(mylist)

```

```python
"""
Decorator THat takes Argument  : Usefull to DIreclty input to the Decorator
Sometimes you want a decorator that behaves differently based on input:
Repeat a function n times,
Delay by x seconds,
Add a custom message,
Only allow access for specific roles,

"""

# repeating n times the function
def repeat(n): # Decorator Factory
    def Decorator(funct):  # Original Decorator
        def wrapper(*args):  # Wrapper Function
            for _ in range(n):
                print("*"*1)
                funct(*args)
        return wrapper
    return Decorator
            
"""
 repeat(3) is not the decorator — its a decorator factory: a function that returns the actual decorator.
 """

@repeat(3)
def myfunction(a,b):
    print(a+b)

myfunction(5,6)
```

```python
# similarly for X seconds delay
import time
def delay(n):   # Decorator Factory to get the argument while @delay(2)
    def Decorator(funct): # Decorator Itself
        def wrapper(*args): # wrpper FUnction to wrap things,like function etc.
            time.sleep(n)
            print("Waiting for ",n,"seconds")
            return funct(*args)
        return wrapper
    return Decorator

@delay(3)
def myfunction():
    print("Hello")

myfunction()
```

Outer Layer	Accepts parameters (like n=3, seconds=2)

Middle Layer	The actual decorator (gets func)

Inner Layer	The wrapper (modifies function behavior)

```python
"""
functools.wraps — Preserving Function Metadata
When you write a decorator, it replaces your original function with the wrapper.

so ,This causes the original functions:

__name__

__doc__

__module__

To be lost ,
SO to presever the original function data we use FUnctools.."""
```

```python
# example
def simple_Decorator(funct):
    def wrapper(*args):
        """this is wrapper """
        return funct(*args)
    return wrapper

@simple_Decorator
def Myfunction(a):
    """ this is my function """
    print(a)

Myfunction(5)
print(Myfunction.__name__)
print(Myfunction.__doc__)
print(Myfunction.__module__)

# we lost both the original function name and the original function docstirng
```

```python
#solution Use functools to overcome this situation as
from functools import wraps

def simple_Decorator(funct):
    @wraps(funct)      #This line fixes everything
    def wrapper(*args):
        """this is wrapper """
        return funct(*args)
    return wrapper

@simple_Decorator
def Myfunction(a):
    """ this is my function """
    print(a)

Myfunction(5)
print(Myfunction.__name__)
print(Myfunction.__doc__)
print(Myfunction.__module__)

"""
What @wraps(func) Actually Does:
It copies:

__name__

__doc__

__annotations__

__module__"""
```

```python
# logging Decorator 
import functools
"""
A log decorator is a wrapper around your function that logs (prints or saves) 
details about the function when it is called.like,

Function name and arguments before execution.
Return value after execution.
Execution time.
Exceptions raised.
"""
def Log_Decorator(funct):
    functools.wraps(funct)
    def wrapper(*args,**kwargs):
        print(f"[Logging To] {funct.__name__} with arguments{args} and Keyword arguments {kwargs}")
        return funct(*args,**kwargs)
    return wrapper

@Log_Decorator
def square(n):
    print(n**2)

@Log_Decorator
def Cube(n):
    print(n**3)

square(5)
Cube(5)
```

```python
"""
Chaining Multiple Decorators
In Python, we can apply more than one decorator
 to a single function. 
 This is called decorator chaining.

 Syntax:
@decorator_one
@decorator_two
def my_func():
    ...

    
    Similar to :
my_func = decorator_one(decorator_two(my_func))

"""

# example
def Decorator1(funct):
    def wrapper(*args):
        print("Decorator 1")
        funct(*args)
    return wrapper


def Decorator2(funct):
    def wrapper(*args):
        print("Decorator 2")
        funct(*args)
    return wrapper

@Decorator1
@Decorator2
def my_function(a,b):
    print(a+b)
# my_function=Decorator1(Decorator2(my_function))

my_function(5,6)

"""
Execution Flow:
1. Decorator2 wraps my_function:
   adds print: "Decorator 2"
   calls my_function(a, b)

2. Decorator1 wraps the result of Decorator2:
   adds print: "Decorator 1"
   calls the result of Decorator2(...)

output:
Decorator 1
Decorator 2
11

Order matters: The bottom decorator applies first.

Each decorator gets the output of the decorator below it.

"""
```

```python
# ADVANCE LOG DECORATOR 
# THAT PRINT THE LOGGING Message
#  and then Save Details to the file

import logging

## Setup logging configuration
logging.basicConfig(
    filename='log.txt', #Log file name
    level=logging.INFO, #
    format='%(asctime)s - %(levelname)s - %(message)s'# message print in Log file
)

def AdvanceLoggingDecorator(funct):
    functools.wraps(funct)
    def wrapper(*args,**kwargs):
        print(f"[Logging To] {funct.__name__} with arguments{args} and Keyword arguments {kwargs}")
        logging.info(f"[Logging To] {funct.__name__} with arguments{args} and Keyword arguments {kwargs}")
        return funct(*args,**kwargs)
    return wrapper
        
@AdvanceLoggingDecorator
def myfunction(A,B,Message):
    print(Message,A+B)

myfunction(5,6,"Addition")

@AdvanceLoggingDecorator
def multi(A,B):
    print(A*B)

multi(5,6)


```

```python
"""
 Stateful Decorators (Function decorators with memory)
These are decorators that store information or state across function calls, such as:

Counting how many times a function is called

Caching previous results"""

def count_decorator(funct):
    functools.wraps(funct)
    def wrapper(*args):
        wrapper.calls+=1
        funct(*args)
        print(f" {funct.__name__} called {wrapper.calls} times")
    wrapper.calls=0
     # here we are creating the attribute of function
     # outside from function as function are the objects in python 
     # we can create the Variables outside the function
    return wrapper

@count_decorator
def hello(name):
     print("Hello,",name)



hello("varun")
hello("prince")
hello("alice")
```

```python
# decorator to check the given data typeto function is right

def checker(datatype):
    def decorator(funct):
        @functools.wraps(funct)
        def wrapper(*args):
            try:  
                if all(isinstance(i,datatype) for i in args[0]):
                    funct(*args)
                else:
                    raise TypeError("Incorrect datatype")
            except Exception as e:
                print(e)
                
        return wrapper
    return decorator

@checker(int)
def square(num):
    for i in num:
       print(i**2)

square([1,2,3])
square([1,2,3,'4'])
```
